```python
import pandas as pd 
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("labeled_data (1).csv")
```

```python
# keys: {0:hate_speech,1:offensive_language,2:neither}
df
```

```python
vectorizer = CountVectorizer()
x = vectorizer.fit_transform(df['tweet'].values.astype('U'))
x = x.toarray()
```

```python
y = df['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
clf = RandomForestClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
round(clf.score(X_test,y_test)*100,2)
```

```python
#keys: {0:hate_speech,1:offensive_language,2:neither}
def predict(str):
    return clf.predict(vectorizer.transform([str]).toarray())
```
